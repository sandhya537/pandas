```python
##----------------adding a new column--------------------------------------

import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print()
print(df.head())
df["status"]="pass"
print()
print(df.head())
```

```python
##----------------------adding a new column avg (with operation)----------------------------

import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df.head())
df["average"]=df["tot"]/3
print()
print(df.head())
```

```python
##------------adding new column with apply()-----------------------------------------
#                                    ======
import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df)
def total(df):
    t=df["tot"]/3
    return t
df["average"]=df.apply(total,axis=1)
print()
print(df.head())
```

```python
##----------------adding new column by insert()---------------------------------
#                                      ========
import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df.head())
new=df["tot"]/3
df.insert(2,"average",new)
print()
print(df)
```

```python
##---------------drop the single column----------------------------------------

import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df.head())
df1=df.drop(columns=["name"],axis=1)
print()
print(df1.head())
```

```python
##--------------------drop the  multiple columns-----------------------------

import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df.head())
df1=df.drop(columns=["name","sid"],axis=1)
print()
print(df1.head())
```

```python
##-------------drop the single row----------------------------------------------------

import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df.head())
df1=df.drop(3,axis=0)
print()
print(df1.head())
```

```python
##-----------------drop the multipul rows----------------------------------------

import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df.head())
df1=df.drop([1,3],axis=0)
print()
print(df1.head())
```

```python
#3-------------modifying specific location value-----------------------

import pandas as pd
df=pd.read_excel("/content/marks sheet1.xlsx")
print(df.head())
df.loc[3,"sid"]=108
print()
print(df.head())
```
