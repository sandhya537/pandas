```python
##----------using isna() and isnull() -----------------------------------------

import pandas as pd
df=pd.read_csv("/content/synthetic_dataset.csv")
df1=df.isna()
print("output is:")
print(df.head())
print()
print("after using isna() :")
print(df1.head())
print()
print("after using isnull() :")
print(df1.head())
```

```python
##------------------using notnull() -------------------------------------------

import pandas as pd
df=pd.read_csv("/content/synthetic_dataset.csv")
df1=df.notnull()
print("output is:")
print(df.head())
print()
print("after using notnull() :")
print(df1.head())
```

```python
##------------------------isna().sum()--------------------------------

import pandas as pd
df=pd.read_csv("/content/synthetic_dataset.csv")
df1=df.isna().sum()
per=(df1*100)/len(df)
print("output is:")
print(df.head())
print()
print("after using isna().sum() :")
print(df1.head())
print()
print(" percentage :")
print(per)
```

```python
##---------------------using dropna()------------------------------------------

import pandas as pd
df=pd.read_csv("/content/synthetic_dataset.csv")
df1=df.dropna()
print("output is:")
print(df.head())
print()
print("after using dropna() :")
print(df1.head())
```

```python
##-------------------------dropna() and isna().sum()-----------------------

import pandas as pd
df=pd.read_csv("/content/synthetic_dataset.csv")
df1=df.dropna()
df2=df1.isna().sum()
print("output is:")
print(df.head())
print()
print("after dropna()and isna().sum() :")
print(df2.head())
```

```python
##--------------creating dataframe,fillna()---------------------------------------------

import pandas as pd
import numpy as np
data=[['lal',34,50000],
      ['bal',23,60000],
      ['pal',45,30000],
      ['tagore',33,np.nan],
      [np.nan,43,60000]
     ]
df=pd.DataFrame(data,columns=['name','age','salary'])
print(df)
print()
df1=df.fillna(0)
print(df1)
```

```python

import pandas as pd
import numpy as np
data=[['ramdas',34,657],
      ['kabirdas',np.nan,876],
      ['tulsidas',23,987],
      ['arundas',33,456],
      [ 'kalidas',np.nan,654]
]
df1=pd.DataFrame(data,columns=['name','age','poetry'])
print(df1)
print()
m=df1['age'].mean()
df1['age']=df1['age'].fillna(m)
print(df1)


```
