```python
##-----------creating dataframe---------------------------------------------------

import pandas as pd
d1={
    "id":[1,2,3,4,5,6],
    "name":['pavan','venu','madhu','neeru','shafi','veeru'],
    "subject":['english','java','html','python','c','dotnet'],
}
d2={
    "id":[11,12,13,14,15,16],
    "name":['srinu','suma','neelu','dhoni','arjun','veeru'],
    "subject":['java','html','cpp','python','c','dotnet'],
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
print(df1)
print()
print(df2)
```

```python
##-----------------------inner merging-----------------------------------------------

import pandas as pd
d1={
    "id":[1,2,3,4,5,6],
    "name":['pavan','venu','madhu','neeru','shafi','veeru'],
    "subject":['english','java','html','python','c','dotnet'],
}
d2={
    "id":[11,12,13,14,15,16],
    "name":['srinu','suma','neelu','dhoni','arjun','veeru'],
    "subject":['java','html','cpp','python','c','dotnet'],
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
inn=pd.merge(df1,df2,on="subject",how="inner")
print(df1)
print()
print(df2)
print()
print(inn)
```

```python
##----------------------------left merging------------------------------------------

import pandas as pd
d1={
    "id":[1,2,3,4,5,6],
    "name":['pavan','venu','madhu','neeru','shafi','veeru'],
    "subject":['english','java','html','python','c','dotnet'],
}
d2={
    "id":[11,12,13,14,15,16],
    "name":['srinu','suma','neelu','dhoni','arjun','veeru'],
    "subject":['java','html','cpp','python','c','dotnet'],
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
left=pd.merge(df1,df2,on="subject",how="left")
print(df1)
print()
print(df2)
print()
print(left)
```

```python
##------------------ right merging-------------------------------------------------

import pandas as pd
d1={
    "id":[1,2,3,4,5,6],
    "name":['pavan','venu','madhu','neeru','shafi','veeru'],
    "subject":['english','java','html','python','c','dotnet'],
}
d2={
    "id":[11,12,13,14,15,16],
    "name":['srinu','suma','neelu','dhoni','arjun','veeru'],
    "subject":['java','html','cpp','python','c','dotnet'],
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
right=pd.merge(df1,df2,on="subject",how="right")
print(df1)
print()
print(df2)
print()
print(right)
```

```python
##------------------outer merging------------------------------------------------

import pandas as pd
d1={
    "id":[1,2,3,4,5,6],
    "name":['pavan','venu','madhu','neeru','shafi','veeru'],
    "subject":['english','java','html','python','c','dotnet'],
}
d2={
    "id":[11,12,13,14,15,16],
    "name":['srinu','suma','neelu','dhoni','arjun','veeru'],
    "subject":['java','html','cpp','python','c','dotnet'],
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
outer=pd.merge(df1,df2,on="subject",how="outer")
print(df1)
print()
print(df2)
print()
print(outer)
```

```python
##-------------one_one merging---------------------------------------

import pandas as pd
d1={
    "employee":["nireekshan","veeru","lavanya","pradhan"],
    "group":["devlopment","testing","testing","hr"]
}
d2={
    "employee":["nireekshan","veeru","lava","pradhan"],
    "hire_date":[2010,2012,2014,2016]
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
one_one=pd.merge(df1,df2)
print(df1)
print(df2)
print()
print(one_one)
```

```python
##---------------many_one------------------------------------------------------------

import pandas as pd
d1={
    "employee":["nireekshan","veeru","lavanya","pradhan"],
    "group":["devlopment","testing","testing","hr"]
}
d2={
    "employee":["nireekshan","veeru","lava","pradhan"],
    "hire_date":[2010,2012,2014,2016]
}
d3={
    "group":["testing","development","hr"],
    "supervisor":["shafi","dhoni","neelima"]
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
df3=pd.DataFrame(d3)
one_one=pd.merge(df1,df2)
many_one=pd.merge(one_one,df3)
print(df1)
print(df2)
print()
print(one_one)
print()
print(many_one)
```
