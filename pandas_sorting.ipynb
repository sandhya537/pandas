```python
##-------------------creating DataFrame-----------------------------------------
               ##--------sort_index------------------------------------

import pandas as pd
d={
    'oid':[11,22,33],
    'cust name':['kedar','varun','tanuja'],
    'product':['iphone 11','htc','macbook']
}
i=[555,444,333]
df1=pd.DataFrame(d,index=i)
df2=df1.sort_index()
print(df1)
print(df2)
```
