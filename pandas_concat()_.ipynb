```python
##------------creating dataframe------------------------------------------------

import pandas as pd
data1=[[11,22],[33,44]]
data2=[[55,66],[77,88]]
columns=['column1','column2']
df1=pd.DataFrame(data1,columns=columns)
df2=pd.DataFrame(data2,columns=columns)
print(df1)
print()
print(df2)
```

```python
##--------------------- using concat()----------------------------------------------------

import pandas as pd
data1=[[11,22],[33,44]]
data2=[[55,66],[77,88]]
columns=['column1','column2']
df1=pd.DataFrame(data1,columns=columns)
df2=pd.DataFrame(data2,columns=columns)
print(df1)
print()
print(df2)
res=[df1,df2]
con=pd.concat(res)
print()
print('after using concat()')
print(con)
```

```python
##-------------------concat(),ignore_index=True------------------------------------

import pandas as pd
data1=[[11,22],[33,44]]
data2=[[55,66],[77,88]]
columns=['column1','column2']
df1=pd.DataFrame(data1,columns=columns)
df2=pd.DataFrame(data2,columns=columns)
print(df1)
print()
print(df2)
res=[df1,df2]
con=pd.concat(res,ignore_index=True)
print()
print('after using concat(),ignore_index=True')
print(con)
```

```python
##---------------------------concat()-----------------------------------------

import pandas as pd
data1=[[12,13,14],[15,16]]
data2=[[17,18],[19,20,21]]
c=['A','B','C']
df1=pd.DataFrame(data1,columns=c)
df2=pd.DataFrame(data2,columns=c)
print(df1)
print()
print(df2)
print()
print('after using concat()')
print()
con=pd.concat([df1,df2])
print(con)

```

```python
##----------------------concat(),ignore_index=True-----------------------------------

import pandas as pd
data1=[[12,13,14],[15,16]]
data2=[[17,18],[19,20,21]]
c=['A','B','C']
df1=pd.DataFrame(data1,columns=c)
df2=pd.DataFrame(data2,columns=c)
print(df1)
print()
print(df2)
print()
print('using concat(),ignore_index=True')
print()
con=pd.concat([df1,df2],ignore_index=True)
print(con)
```

```python
##--------------------concat(),axis=1-----------------------------------------------

import pandas as pd
data1=[[10,20],[40,50,65]]
data2=[[60,70,80],[90,100]]
columns=['column1','column2','column3']
df1=pd.DataFrame(data1,columns=columns)
df2=pd.DataFrame(data2,columns=columns)
print(df1)
print()
print(df2)
print()
print('using concat(),axis=1 ')
print()
con=pd.concat([df1,df2],axis=1)
print(con)
```

```python
##---------------------concat(),axis=1------------------------------------

import pandas as pd
data1=[[10,20],[40,50,65]]
data2=[[60,70,80],[90,100]]
columns=['column1','column2','column3']
df1=pd.DataFrame(data1,columns=columns)
df2=pd.DataFrame(data2,columns=columns)
print(df1)
print()
print(df2)
print()
print('using concat(),axis=1')
print()
con=pd.concat([df1,df2],axis=1)
con.columns=[f'column{i+1}' for i in range(con.shape[1])]
print(con)
```
