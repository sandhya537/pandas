```python
##--------creating DF with missing date------------------------------------

import pandas as pd
data={
    "product":["samsung","iphone","motorola"],
    "status":["success","success","failed"],
    "cost":[10000,50000,15000],
    "purdate":["02-sep-2019","date is missing","21-sep-2019"]
}
df=pd.DataFrame(data)
print(df)
print()
print(df.dtypes)


```

```python
##------------using errors="coerce"---------------------------------------
import pandas as pd
data={
    "product":["samsung","iphone","motorola"],
    "status":["success","success","failed"],
    "cost":[10000,50000,15000],
    "purdate":["02-sep-2019","date is missing","21-sep-2019"]
}
df=pd.DataFrame(data)
df["purdate"]=pd.to_datetime(df["purdate"],errors="coerce")
print(df)
print()
print(df.dtypes)
```

```python
##--------------displaying perticular month ,day and year data----------------------------------------

import pandas as pd
df=pd.read_csv("/content/sales7_dates.csv",parse_dates=['Pur_Date'])

df["year"]=df["Pur_Date"].dt.year
df["month"]=df["Pur_Date"].dt.month
df["day"]=df["Pur_Date"].dt.day
print(df.head())
```

```python
##-------- displaying perticular min ,second ,hours data-----------------------------------------

import pandas as pd
df=pd.read_csv("/content/sales7_dates.csv",parse_dates=['Pur_Date'])

df["year"]=df["Pur_Date"].dt.year
df["month"]=df["Pur_Date"].dt.month
df["day"]=df["Pur_Date"].dt.day
df["hour"]=df["Pur_Date"].dt.hour
df["minute"]=df["Pur_Date"].dt.minute
df["second"]=df["Pur_Date"].dt.second
print(df.head())
```

```python
##-----------displaying daynames of dates---------------------------------------------

import pandas as pd
data={
    "product":["samsung","iphone","motorola"],
    "status":["success","success","failed"],
    "cost":[10000,50000,15000],
    "purdate":["2018-01-01","2018-01-02","2018-01-03"]
}
df=pd.DataFrame(data)
df['purdate']=pd.to_datetime(df['purdate'])
print(df)
print()
print(df['purdate'].dt.day_name())
```

```python
##--------------------displaying weeks count of dates ----------------------------------------

import pandas as pd
data={
    "product":["samsung","iphone","motorola"],
    "status":["success","success","failed"],
    "cost":[10000,50000,15000],
    "purdate":["2018-01-01","2018-01-02","2018-01-03"]
}
df=pd.DataFrame(data)
df['purdate']=pd.to_datetime(df['purdate'])
print(df)
print()
print(df['purdate'].dt.weekday)
```
